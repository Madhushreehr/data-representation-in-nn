```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# 1. Load Dataset
# num_words=5000 matches your embedding layer input dimension
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)

# 2. Preprocess / Padding
# Standardizing sequence length to 100 as per your notes
maxlen = 100
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# 3. Build Transformer-based Model
inputs = tf.keras.Input(shape=(maxlen,))
x = layers.Embedding(input_dim=5000, output_dim=32)(inputs)

# Multi-Head Attention layer
# Note: In a full transformer, you'd add Positional Encoding here
x = layers.MultiHeadAttention(num_heads=2, key_dim=32)(x, x)
x = layers.GlobalAveragePooling1D()(x)

# Dense Layers
x = layers.Dense(10, activation='relu')(x)
outputs = layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

# 4. Compile Model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 5. Train Model
print("Starting training...")
model.fit(x_train, y_train, epochs=3, batch_size=32, validation_split=0.2)

# 6. Evaluate and Print Results
print("\nEvaluating on test data:")
loss, acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {acc:.4f}")
```
